```python
#melakukan normalisasi

def norm(df):
    df = (df-df.min())/(df.max()-df.min())
    return df
```

```python
#menghitung euclidean
def euclidean(x1,x2):
    return np.sqrt(np.sum(x1-x2)**2)

```

```python
#training KNN
def knn(x_train, y_train, x_test,k):
    dist = []

    #menghitung distance dari data trainining dan data testing
    for row in range(x_train.shape[0]):
        dist.append(euclidean(x_train.iloc[row],x_test))
    
    #menambahkan dara distance pada data
    data = x_train.copy()
    data["Dist"] = dist

    #menambahkan class pada data
    data["Class"] = y_train

    #mengurutkan data berdasarkan distance
    data = data.sort_values(by="Dist").reset_index(drop=True)

    #mengambil label kelas yang paling sering muncul di kelas
    y_pred = data.iloc[:k].Class.mode()

    return y_pred[0]


    
```

EVALUATION

```python
#menghitung akurasi dari output  berdasarkan label kelas
def acc(y_pred,y_true):
    benar = 0 
    for i in range(len(y_pred)):
        print(y_pred[i] == y_true[i])
        if y_pred[i] == y_true[i]:
            benar+=1
            
    return benar/len(y_pred)


```

```python
#evaluasi model dengan menggunakan data fold

def evaluate(fold, k):
    test,train = fold
    x_train,ytrain = train.drop("Species",axis=1),train.Species
    x_test,ytest = test.drop("Species",axis=1),test.Species
    

    x_train = norm(x_train)
    x_test = norm(x_test)
    y_pred = []
    for row in range(x_test.shape[0]):
        y_pred.append(knn(x_train,ytrain,x_test.iloc[row],k))
    

    return (acc(y_pred,ytest))



```

```python
#main Program
k = 5
accs = []
folds = [data_fold1, data_fold2, data_fold3]
for i in range(len(folds)):
    accs.append(evaluate(folds[i],k))

print(f'Menggunakan k:{k}, dengan rata-rata akurasi: {sum(accs)/3}')
```
